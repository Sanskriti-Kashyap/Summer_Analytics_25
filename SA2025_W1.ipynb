### 1. What is name of car that has the highest horsepower?

### 2. How many cars have mpg  35?

```python
#your code here
# Count cars with mpg >= 35
count_mpg_35_or_more = (df['mpg'] >= 35).sum()
print(count_mpg_35_or_more)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
# Filter the dataframe
filtered_df = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_df['origin'].mode()[0]

print(most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
# Filter cars from Japan (case-insensitive)
japan_cars = df[df['origin'].str.lower() == 'japan']

# Calculate mean acceleration and round to 2 decimals
mean_acceleration = round(japan_cars['acceleration'].mean(), 2)

print(mean_acceleration)
```

### 5. Which year had the highest average mpg?

```python
#your code here
# Group by year and calculate mean mpg
mean_mpg_by_year = df.groupby('model_year')['mpg'].mean()

# Find the year with the highest average mpg
year_with_highest_avg_mpg = mean_mpg_by_year.idxmax()
highest_avg_mpg = mean_mpg_by_year.max()

print(f"Year with highest average mpg: {year_with_highest_avg_mpg} ({highest_avg_mpg:.2f} mpg)")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average mpg per year per origin
avg_mpg_by_year_origin = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot using Seaborn lineplot
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG Evolution Over Years by Origin')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))

sns.scatterplot(
    data=df.reset_index(),  # just in case index isn't default
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),          # size range for mpg bubbles
    hue_order=['japan', 'europe', 'usa'],
    palette='Set2',
    alpha=0.7,
    edgecolor='w',
    linewidth=0.7
)

plt.title('Car Weight vs Horsepower Colored by Origin and Sized by MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight', fontsize=14)
plt.legend(title='Origin', loc='upper right', fontsize=12, title_fontsize=13)

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
